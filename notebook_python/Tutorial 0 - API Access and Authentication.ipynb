# Tutorial 0 - API Access and Authentication

Before we start the analysis, we need to connect to the ParlaCAP API.

There are two docs pages:

- `https://parlacap.ipipan.waw.pl/auth/docs#/` is for user access: register, login, token, and API keys.
- `https://parlacap.ipipan.waw.pl/api/v1/docs#/` is the main data API: this is where you use the API key to query speeches.

The login endpoint docs are here:

- `https://parlacap.ipipan.waw.pl/auth/docs#/User%20Panel/login_token_post`

## Short setup flow

1. Open the auth docs and create an account with `POST /register`.
2. Log in with `POST /token`. This gives you an `access_token`.
3. In the auth docs click **Authorize** and paste the bearer token there.
4. While still authenticated, use `POST /me/keys` to create an API key.
5. Copy the API key right away. It is shown only once.
6. Open the backend docs and use **Authorize** again. Paste the API key into `APIKeyHeader (apiKey)`.
7. Save the same key in `my_secrets.env` one folder above this notebook.

Your file should look like this:

`PARLACAP_API_KEY=YOUR_KEY_HERE`

If you run the code cell below and get one row back, the setup works.


## Before you run this notebook

This notebook is only meant to check one thing: whether your API access works. You do **not** need to understand the whole dataset yet.

You will need:
- a working Python environment
- a `my_secrets.env` file in the repo root
- an API key stored as `PARLACAP_API_KEY=...`

## What success looks like

If the code cell below returns one row of data, your setup is working and you can continue to Tutorial 1.

## If something fails

- a `401` or `403` error usually means the API key is missing, invalid, or not authorized
- a file error usually means `my_secrets.env` is not in the repo root
- an import error usually means the environment is not installed yet

````python
import os
import requests

# Read the key from my_secrets.env.
os.environ["PARLACAP_API_KEY"] = open("../my_secrets.env", encoding="utf-8").read().split("=", 1)[1].strip()
headers = {"X-API-Key": os.getenv("PARLACAP_API_KEY")}

# Ask the API for one row.
query = {
    "limit": 1,
    "offset": 0,
    "columns": ["date", "cap_category", "parliament"],
}

response = requests.post("https://parlacap.ipipan.waw.pl/api/v1/filter", json=query, headers=headers)
response.raise_for_status()
response.json()["rows"][0]
````

If you see one row of data, everything works and you can open Tutorial 1.


## Preparing a new Python environment

If you want to replicate the output of these notebooks exactly, it is strongly recommended to create a fresh environment with the library versions used in this tutorial.

There are two easy ways to do this.

### Using conda or mamba

If you use conda or mamba, you can create the environment from the `env.yaml` file in this repository:

```bash
mamba env create -n parlacaptutorials -f env.yaml
```

Then select `parlacaptutorials` as your notebook interpreter.

### Using venv

First, make sure you have Python 3.14.2 installed.

Create a new environment with:

```bash
python3.14 -m venv parlacaptutorials
```

Activate it with:

```bash
source parlacaptutorials/bin/activate
```

If you use Windows, use this instead:

```powershell
parlacaptutorials\Scripts\activate
```

Then install the required libraries:

```bash
pip install -r requirements.txt
```
